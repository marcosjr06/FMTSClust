```python
# STATION 1

import pandas as pd

data = pd.read_csv("./data/SIM001-FULL.txt", delimiter = "\t", header=None)

#get specific columns
df_filtered = data.iloc[:, [0,1,2,5,17,19]]
df_filtered.columns = ['Date', 'Time', 'Temp', 'RelHum', 'Rain', 'SolarRad']
df_result = df_filtered.iloc[2:]
df_result.reset_index(drop=True, inplace=True)
df_result.insert(0, "StationID", 1, True) # add Station ID

df_result.to_csv('./data_csv/Station01.csv', index=False)
df_result
```
